# 2. Chương trình chính thức

```python
lst_so_ky_hieu = []
lst_ngay_ban_hanh = []
lst_trich_yeu = []
lst_hieu_luc = []
lst_pdf_links = []
```

```python
clicking_time = 354

# create a chrome driver 
# PATH = r"C:\Users\AD\Downloads\chromedriver.exe"
# driver = webdriver.Chrome(PATH)

# # get url 
# url = 'https://www.binhduong.gov.vn/chinh-quyen/van-ban-chi-dao-dieu-hanh'
# driver.get(url)
# time.sleep(20)

for i in range(1,clicking_time):

    check_num_page = int(re.search('\d',driver.find_element(by= By.CSS_SELECTOR,value = 'li > a.pageLink[title="Trang sau"]').get_attribute('href')).group())
    print('Đang ở trang số {}'.format(check_num_page))
    print('----------------------------------------')

    # Lấy số ký hiệu
    so_ky_hieu_elements = driver.find_elements(by= By.XPATH, value='//*[@id="ctl00_ctl33_g_bd312eca_ef7a_4d42_b030_d136358de9de_ctl00_Result"]/table/tbody/tr/td[2]/a')
    so_ky_hieu = [x.text for x in so_ky_hieu_elements]
    lst_so_ky_hieu.append(so_ky_hieu)
    print(so_ky_hieu[:2])
    print('----------------------------------------')

    # Lấy ngày ban hành 
    ngay_ban_hanh_elements = driver.find_elements(by= By.XPATH,value='//*[@id="ctl00_ctl33_g_bd312eca_ef7a_4d42_b030_d136358de9de_ctl00_Result"]/table/tbody/tr/td[3]/a')
    ngay_ban_hanh = [x.text for x in ngay_ban_hanh_elements]
    lst_ngay_ban_hanh.append(ngay_ban_hanh)
    print(ngay_ban_hanh[:2])
    print('----------------------------------------')
    
    # Lấy trích yếu 
    trich_yeu_elements = driver.find_elements(by= By.XPATH,value='//*[@id="ctl00_ctl33_g_bd312eca_ef7a_4d42_b030_d136358de9de_ctl00_Result"]/table/tbody/tr/td[4]/a/div')
    trich_yeu = [x.text for x in trich_yeu_elements]
    lst_trich_yeu.append(trich_yeu)
    print(trich_yeu[:2])
    print('----------------------------------------')

    # Lấy hiệu lực 
    hieu_luc_elements = driver.find_elements(by= By.XPATH,value='//*[@id="ctl00_ctl33_g_bd312eca_ef7a_4d42_b030_d136358de9de_ctl00_Result"]/table/tbody/tr/td[5]/a')
    hieu_luc = [x.text for x in hieu_luc_elements]
    lst_hieu_luc.append(hieu_luc)
    print(hieu_luc[:2])
    print('----------------------------------------')


    # Lấy link pdf
    pdf_elements = driver.find_elements(by= By.XPATH,value='//*[@id="ctl00_ctl33_g_bd312eca_ef7a_4d42_b030_d136358de9de_ctl00_Result"]/table/tbody/tr/td[6]/div/a')
    pdf_links = [x.get_attribute('href') for x in pdf_elements]
    lst_pdf_links.append(pdf_links)
    print(pdf_links[:2])
    print('----------------------------------------')

    cick = driver.find_element(by = By.CSS_SELECTOR,value='li > a.pageLink[title="Trang sau"]').click()  

    check_num_page = int(re.search('\d',driver.find_element(by= By.CSS_SELECTOR,value = 'li > a.pageLink[title="Trang sau"]').get_attribute('href')).group())

    if check_num_page:
        print('Đã qua trang số {}'.format(check_num_page))
        print('*******'.center(30,'-'))
        print(len(lst_so_ky_hieu),len(lst_ngay_ban_hanh),len(lst_pdf_links),len(lst_trich_yeu), len(lst_hieu_luc),end='')
```

```python
print('Ngày ban hành: ',len(lst_ngay_ban_hanh))
print('Ngày hiệu lực: ', len(lst_hieu_luc))
print('Số ký hiệu: ',len(lst_so_ky_hieu))
print('Trích yếu: ',len(lst_trich_yeu))
print('PDF: ',len(lst_pdf_links))
```

```python
def join_lst(old_lst,new_lst):
    for lst in old_lst:
        for val in lst:
            new_lst.append(val)
```

```python
for lst in lst_pdf_links:
    if len(lst) != 12:
        print('lỗi ở list số: {}'.format(lst_pdf_links.index(lst)))
```

```python
for lst in lst_trich_yeu:
    if len(lst) != 12:
        print('lỗi ở list số: {}'.format(lst_trich_yeu.index(lst)))
```

```python
lst_index_can_xoa = [17,32-1,33-2,34-3,44-4,77-5,78-6,98-7,104-8,114-9,186-10,196-11,214-12,256-13,294-14,350-15,351-16,352-17]
def xoa_lst(old_lst):
    for i in lst_index_can_xoa:
        del old_lst[i]
```

```python
xoa_lst(lst_hieu_luc)
xoa_lst(lst_ngay_ban_hanh)
xoa_lst(lst_so_ky_hieu)
xoa_lst(lst_trich_yeu)
xoa_lst(lst_pdf_links)
```

```python
lst_so_ky_hieu_new = []
lst_ngay_ban_hanh_new = []
lst_hieu_luc_new = []
lst_trich_yeu_new = []
lst_pdf_links_new = []
```

```python
join_lst(lst_hieu_luc,lst_hieu_luc_new)
join_lst(lst_ngay_ban_hanh,lst_ngay_ban_hanh_new)
join_lst(lst_so_ky_hieu,lst_so_ky_hieu_new)
join_lst(lst_trich_yeu,lst_trich_yeu_new)
join_lst(lst_pdf_links,lst_pdf_links_new)
```
